## Expressões Lógicas

Da mesma forma que se pode fazer operações em valores numéricos, existem operadores para valores booleanos.

Os principais são:

- not
    - O comando not vai inverter o booleano. `not True = False` e vice versa.
- and
    - O comando and é uma comparação de E lógica. Ela funciona quando você quer que ambas as condições sejam verdadeiras. Sendo assim:
        - True and True = True -> Porque ambas são verdadeiras
        - True and False = False -> Porque não são todas verdadeiras
        - False and True = False -> Porque não são todas verdadeiras
        - False and False = False -> Porque não são todas verdadeiras
- or
    -  O comando or é uma comparação do OU lógica. Ela funciona quando você quer que qualquer uma das condições seja verdadeira.
        - True or True = True -> Porque ao menos uma é verdadeira
        - True or False = True -> Porque ao menos uma é verdadeira
        - False or True = True -> Porque ao menos uma é verdadeira
        - False or False = False -> Porque nenhuma é verdadeira
    

````python
True and True
````

````python
True and False
````

````python
True or False
````

````python
(3 > 1) and (3 > 2)
````

````python
(3 > 1) and (1 > 3)
````

````python
almocei = True
jantei = False

almocei and jantei
````

## Condicionais

Um dos comandos mais importantes da programação são os comandos condicionais. Eles servem para seu programar decidir se irá fazer alguma coisa, ou não.

Condições são conceitos muito presentes em nossas vidas. Por exemplo, você pode estar pensando se vai almoçar e pensar:

_Se estou com fome, vou almoçar_

Na programação a mesma coisa pode ser feita com o comando <b>if</b>. Você pode criar uma condição com o if, e o bloco de código que estiver dentro do if só será executado caso aquela condição seja verdadeira.

```python
if condição_do_if:
    # Bloco de código
    # do if. Só será executado
    # se a condição for verdadeira

# Fora do if -> não depende da condição
```

**Obs:** Atenção com a identação! O que define o que está dentro da condição é a identação do bloco de código

**Obs 2:** Atenção com os `:` no final da linha do if

````python
estou_com_fome = True

if estou_com_fome:
    print('Vou almoçar!')
````

### Exemplos:

1) Pergunte a idade de uma pessoa. Se ela for maior de idade, diga que você pode vender bebida para ela. Se não for, diga que você não pode. De qualquer jeito, agradeça no final!

````python
idade = int(input('Qual é a sua idade?'))

eh_maior_de_idade = idade > 18

if eh_maior_de_idade:
    print('Quer uma cerveja?')
if not eh_maior_de_idade:
    print(f'Você não pode beber álcool porque só tem {idade} anos. Quer um guaraná?')

print('Obrigado e volte sempre!')
````

### Comando else

Para facilitar o exemplo acima, existe o comando **else**. Você, deve utilizá-lo sempre que não quiser uma condição. Seria equivalente ao se não da nossa linguagem falada:

Se você for maior de idade, vou te vender uma cerveja, se não, vou pedir para tomar um guaraná.

````python
idade = int(input('Qual é a sua idade?'))

if idade > 18:
    print('Quer uma cerveja?')
else:
    print('Você não pode beber álcool. Quer um guaraná?')

print('Obrigado e volte sempre!')
````

### Comando elif

Por fim, você ainda pode alinhar mais condições com o comando elif (equivalente ao um else + um if).

Por exemplo, vamos supor que você quer verificar se a pessoa tem idade mínima para vender guaraná também (se a pessoa for um bebe, ela não pode nem isso).

Isso poderia ser feito com ifs alinhados, da seguinte maneira:

````python
idade = int(input("Qual é a sua idade? "))

if idade > 18:
    print("Quer uma cerveja?")
else:
    if idade > 2:
        print("Quer um refrigerante?")
    else:
        print("Não pode beber refrigerate. Fale com sua mãe.")
````

Porém para a leitura ficar mais clara, é recomendável utilizar o comando elif que já junta as 2 coisas

````python
idade = int(input('Qual é a sua idade?'))

if idade > 18:
    print('Quer uma cerveja?')
elif idade > 2:
    print('Quer um guaraná')
else:
    print('Você é um bebê. Quer um leite?')
````

Escreva um programa que simula um radar eletrônico cuja velocidade máxima é 80 km/h. Receba de input a velocidade de um carro.

### Exercício 3 - Analisador de Variantes Genéticas.

Você está analisando uma variante genética e precisa saber se ela é relevante para análise ou não.

Obs: Esse exercício é uma simplificação! Depois vocês vão saber avaliar de verdade!

Você vai receber 5 parametros:

Frequência Populacional: Frequencia da variante na população em porcentagem.

Gene: Gene da variante.

Impacto: Se ela está numa posição de impacto ALTO ou BAIXO.

Reads: Quantidade de reads da variante.

Vaf: Frequencia alélica da variante.

A primeira coisa é tomar cuidado com a qualidade da leitura. Se a variante tiver menos de 10 reads OU uma frequência alélica abaixo de 20% a gente vai dizer que ela não é relevante, pois deve ser um artefato.

Se ela não for um artefato temos que avaliar as seguintes coisas:

1) Ela só vai ser relevante se o impacto for ALTO.
2) Ela não vai ser relevante se a frequência dela for menor que 5%, a não ser que esteja nos seguintes genes de exceção: 'HFE', 'MEFV' ou 'GJB2'.

````python
frequencia_populacional = float(input('Digite a frequencia populacional (em porcentagem):'))
gene = input('Digite o gene:')
impacto = input('Digite a Impacto (ALTO ou BAIXO):')
reads = float(input('Digite os reads:'))
frequencia_alelica = float(input('Digite a frequencia alélica (em porcentagem):'))

eh_artefato = (frequencia_alelica < 20) or (reads < 10)
impacto_baixo = impacto == 'BAIXO'
esta_em_genes_de_excecao = (gene == 'HFE') or (gene == 'MEFV') or (gene == 'GJB2')
frequencia_populacional_alta = frequencia_populacional > 5

if eh_artefato:
    print('Não é relevante. Deve ser artefato.')
elif impacto_baixo or (frequencia_populacional_alta and not esta_em_genes_de_excecao):
    print('Não é relevante.')
else:
    print('É relevante. Melhor olhar com calma!')
````

## Comandos de repetição

Muitas vezes nós não queremos executar um trecho de código apenas uma vez, mas sim várias vezes.

Para isso, utilizamos comandos de repetições, como `for` e `while`

### While

O comando `while` permite que um código seja executado enquanto uma condição é verdadeira.

```python
while condicao:
    # Código dentro do while vai ser executado enquanto a condição for verdadeira.
```

#### Exemplo 1

Vamos supor que você quer fazer um programa que recebe um número de 0 a 10. 

Caso o usuário digite um número fora do valor mencionado, você pede para ele tentar novamente. 

Isso pode ser feito da seguinte forma:

````python
numero_esta_certo = False
while not numero_esta_certo:
    numero = int(input('Digite um número de 0 a 10: '))
    if numero >= 0 and numero <= 10:
        numero_esta_certo = True
    else:
        print('Você digitou um número errado!')

print('Você digitou o número', numero)
````

#### Exemplo 2

Utilizando apenas o comando while, imprima na tela a tabuada de um número:

Exemplo:

Digite um número: 2

Resposta:

2 x 1 = 2

2 x 2 = 4

2 x 3 = 6

...

````python
numero = int(input('Digite um número: '))

i = 1
while i <= 10:
    print(f'{numero} x {i} = {numero * i}')
    i = i + 1
````

### For / range

Outro comando muito útil é o comando `for`.

Você pode utilizá-lo para fazer uma tarefa  um determinado número de vezes e é mais direto que usar o comando while.

Então, para fazer um processo 10x, pode-se fazer da seguinte forma:

````python
for i in range(10):
    print(i)
````

O comando range por padrão, `range(n)` percorre os valores de de 0 a n-1. Ou seja, range(10) vai de 0 a 9.

Porém isso é customizável.

Ele pode receber até 3 argumentos. Caso ele os receba, o primeiro é o valor de start, o segundo é o valor até onde ele vai percorrer (não incluso ele mesmo), e por fim o delta que ele vai incrementar em cada iteração.

Por exemplo `range(2, 20, 2)` vai percorrer os valores de 2 a 20 (não incluindo o 20) de 2 em 2.

````python
for i in range(2, 20, 2):
    print(i)
````

#### Exemplo 1

Imprima na tela a tabuada de um número, agora usando o comando `for`:

Exemplo:

Digite um número: 2

Resposta:

2 x 1 = 2

2 x 2 = 4

2 x 3 = 6

...

````python
numero = int(input('Digite um número: '))

for i in range(1, 11):
    print(f'{numero} x {i} = {numero * i}')
````

#### Exemplo 2

Pergunte a idade de 5 pessoas, e no fim imprima a maior e menor idade

Fonte: Curso em vídeo

````python
maior_idade = 0
menor_idade = 0

for i in range(5):
    idade = int(input('Digite a idade: '))
    
    eh_a_primeira_repeticao = (i == 0)
    
    if eh_a_primeira_repeticao:
        maior_idade = idade
        menor_idade = idade
    else:
        if idade > maior_idade:
            maior_idade = idade
        if idade < menor_idade:
            menor_idade = idade

print(f'A maior idade é: {maior_idade}')
print(f'A menor idade é: {menor_idade}')
````

#### Exemplo 3

Ex:

Dica: Utilize mais de um for!

#### Exemplo 1 - Achando a base complementar

Repita o exercício anterior de achar a base complementar. Porém, agora você não vai receber apenas uma base, mas sim uma string com tamanho qualquer de pares de bases.

Ex:

Digite a sequencia: ATATTC

Resposta:
TATAAG

**Dica**: Não se esqueça que você pode concatenar strings "somando" elas

````python
sequencia_inicial = input('Digite a sequência: ')
sequencia_final = ''

for base in sequencia_inicial:
    if base == 'A':
        sequencia_final += 'T'
    elif base == 'T':
        sequencia_final += 'A'
    elif base == 'G':
        sequencia_final += 'C'
    elif base == 'C':
        sequencia_final += 'G'

print(sequencia_final)
````
